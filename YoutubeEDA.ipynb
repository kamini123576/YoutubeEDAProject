6. Data Cleaning  (Replace '--'  to NaN)

Dropping the Missing Values

```python
data.dropna(axis = 0, inplace = True)
```

8. Data Cleaning [ Rank Column ]

```python
# Removing the comma and string in the Rank Column
data.dtypes
```

```python
# Removing the string and the comma in the Rank Column
data['Rank'] = data['Rank'].str.replace(r'[^\d]', '', regex=True)
data.tail()
```

```python
# Changing the Datatype of Rank Column
data['Rank'] = data['Rank'].astype(int)
data.dtypes
```

9. Data Cleaning [ Video Uploads & Subscribers ]

```python
data['Subscribers'] = data['Subscribers'].astype('int')
data['Video Uploads'] = data['Video Uploads'].astype('int')
data.dtypes
```

10. Data Cleaning [ Grade Column ]

```python
# For Different Grading Values we are mapping into Numeric Values
data['Grade'].unique()
```

```python
data['Grade'] = data['Grade'].map({'A++ ': 5, 'A+ ' : 4, 'A ' :3, 'A- ' : 2, 'B+ ' :1})
```

```python
data.head()
```

```python
data.dtypes
```
